### 4.1 Решите систему уравнений: y = x2 – 1 и exp(x) + x∙(1 – y) = 1

```python
plt.figure(figsize=(10,10))
plt.axis([-10, 30, -2.5, 30])

x = np.linspace(-10 * np.pi, 10 * np.pi, 1000)

plt.plot(x, x**2 -1)
plt.plot(x, 1 - (1 - np.exp(x))/x)


plt.grid(color='lightgray', linestyle='--')
plt.show()
```

```python
def f(xy):
    x, y = xy
    return(x ** 2 - y - 1, np.exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(f, (2, 5))
print(f'x1 = {x1}, y1 = {y1}')
```

### 4.2 y = x2 – 1 и exp(x) + x∙(1 – y) > 1

```python
def f(xy):
    x, y = xy
    return(x ** 2 - y - 1, np.exp(x) + x * (1 - y) - 1 - 1)

x1, y1 = fsolve(f, (2, 5))
print(f'x1 = {x1}, y1 = {y1}')
```
